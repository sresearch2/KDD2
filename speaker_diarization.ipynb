```python
def get_speaker_seperated_script(df, speaker_timings):
    script = ""
    past_speaker = ""
    for start, end, speaker in speaker_timings:
        if past_speaker != speaker:
            script += f"\n{speaker}:"
        past_speaker = speaker
        script += " ".join(
            [x.strip() for x in df[df["start"] > start][df["end"] < end].word.values]
        )
    return script


from functools import lru_cache


@lru_cache(maxsize=None)
def get_script(video):
    audio = librosa.load(video, sr=16000)
    audio_in_memory = {
        "waveform": torch.tensor(audio[0]).unsqueeze(0),
        "sample_rate": 16000,
    }
    transcript = model.transcribe(word_timestamps=True, audio=audio[0])
    diarization = pipeline(audio_in_memory)
    speaker_timings = []
    for turn, _, speaker in diarization.itertracks(yield_label=True):
        speaker_timings.append([turn.start, turn.end, speaker])
    df = pd.DataFrame([word for x in transcript["segments"] for word in x["words"]])
    script = get_speaker_seperated_script(df, speaker_timings)
    return script
```
